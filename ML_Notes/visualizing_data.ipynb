```python
import matplotlib.pyplot as plt 

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]


#plot() will create a line chart
plt.plot(years,gdp, color='blue', marker='*') #we can plot our chart using this

plt.title("GDP Chart 1950-2010") #we can give a title to chart

plt.ylabel('Billion of $') #y label 


```
