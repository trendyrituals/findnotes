```python
import matplotlib.pyplot as plt 

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]


#plot() will create a line chart
plt.plot(years,gdp, color='blue', marker='*') #we can plot our chart using this

plt.title("GDP Line Chart 1950-2010") #we can give a title to chart

plt.ylabel('Billion of $') #y label 


```

```python
# we can use bar char to know how the values are distributed


from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]


roundfigure = lambda grade : grade//10*10  # its like 83//10*10

histogram = Counter(roundfigure(grade) for grade in grades) #counter will create a dict

plt.bar(histogram.keys(),histogram.values(),5)

# type(histogram)
# Counter?
histogram
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

total_error = [x+y for x,y in zip(variance,bias_squared)] #MSE

xs = [i for i,_ in enumerate(variance)] #gives us token

plt.plot(xs,variance,'g-',label='variance')
plt.plot(xs,bias_squared,'b:', label='bias^2')
plt.plot(xs,total_error,'r-.', label='total error')

plt.legend(loc=9) # show us the labels in our chart
plt.xlabel("model complexity")
plt.title("The Bias-Variance Tradeoff")
```

```python
# A scatterplot is the right choice for visualizing
# the relationship between two paired sets of data.

friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends,minutes)

for label,friends,minutes in zip(labels,friends,minutes):
    plt.annotate(label,xy=(friends,minutes),
                 xytext=(5,-5),
                 textcoords='offset points'
                )
    
plt.title('Friends Group Vs Daily Time Spend')
```
